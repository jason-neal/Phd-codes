```python
Save_dir = "/home/jneal/Phd/Codes/Phd-codes/Notebooks/DRACS-testing/images"

def plot_polynomial_and_order_nods(star="HD167665-1a", detector=1, funct="legendre", order=1, norm=False, nod="all", show=True, save=False):
    """Function to print out the nods for the 
    
    """
    data_path = "/home/jneal/Phd/data/Crires/BDs-DRACS/{}-copy_to_find_artifacts/".format(star)
    folder = "{}_o{}".format(funct, order)
    this_folder = os.path.join(data_path, folder) + "/"
    
#    if norm:
#        file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.norm.fits")
#    else:
#        file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.fits")
#   
#    if nod == "all":
#        pass
#    elif nod in [12345678]:
#        file_names = file_names[nod - 1]
#    else:
#        raise 
#        
#    print(file_names)
#    
#    file_data = []
#    for name in file_names:
#        this_file = os.path.join(this_folder, name)
#        data = fits.getdata(this_file)
#            #print(data)
#        if len(data) == 3:
#            file_data.append(data[0][0])
#        else:
#            file_data.append(data)
#
#    for nod, data in enumerate(file_data):
#          plt.plot(data, label="nod " + str(nod+1))
#    plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
#    plt.title("{}, detector={}, trace={}, order={}".format(star, detector, funct, order))
#    plt.show()
    
    norm_file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.norm.fits")
    file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.fits")
   
    if nod == "all":
        pass
    elif nod in [12345678]:
        file_names = file_names[nod - 1]
        norm_file_names = norm_file_names[nod - 1]
    else:
        raise 
        
    #print(file_names)
    
    file_data = []
    for name in file_names:
        this_file = os.path.join(this_folder, name)
        data = fits.getdata(this_file)
            #print(data)
        if len(data) == 3:
            file_data.append(data[0][0])
        else:
            file_data.append(data)
    norm_file_data = []
    for name in norm_file_names:
        norm_file = os.path.join(this_folder, name)
        norm_data = fits.getdata(norm_file)
        if len(norm_data) == 3:
            norm_file_data.append(norm_data[0][0])
        else:
            norm_file_data.append(norm_data)
    
    fig = plt.figure()
    fig.subplots_adjust(hspace=0.5)
    plt.subplot(211)
    for nod, data in enumerate(file_data):
        plt.plot(data, label="nod " + str(nod+1))
        plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
    plt.title("{}, chip={}, function={}, order={}".format(star, detector, funct, order))
    plt.xlim([0,1024])
    
    plt.subplot(212)
    for nod, norm_data in enumerate(norm_file_data):
          plt.plot(norm_data, label="nod " + str(nod+1))
    #plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
    plt.title("Normalized {}, chip={}, function={}, order={}".format(star, detector, funct, order))
    plt.xlim([0,1024])
    
    if save:
        save_name = "poly_cycle-a_{0}_chip_{1}_{2}_o{3}.".format(star, detector, funct, order)
        plt.savefig(os.path.join(Save_dir, save_name + '.png'), bbox_inches='tight')
        plt.savefig(os.path.join(Save_dir, save_name + '.pdf'), bbox_inches='tight')
    
    if show:
        plt.show()
    else:
        plt.close()
        plt.close(fig)
        
def plot_functions_for_each_nod(star="HD167665-1a", detector=1, order=1, norm=False, nod=0, show=True, save=False):
    """Function to print out the nods for the 
    
    """
    functions = ["spline1", "spline3", "legendre", "chebyshev"]
    
    data_path = "/home/jneal/Phd/data/Crires/BDs-DRACS/{}-copy_to_find_artifacts/".format(star)
    fig = plt.figure()
    fig.subplots_adjust(hspace=0.5)
    for funct in functions:
        folder = "{}_o{}".format(funct, order)
        this_folder = os.path.join(data_path, folder) + "/"
#        if norm:
#            file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.norm.fits")   
#        else:
#            file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.fits")
#    
#        this_nod = file_names[nod]
#    
#        this_file = os.path.join(this_folder, this_nod)
#        data = fits.getdata(this_file)
#        if len(data) == 3:
#            data = data[0][0]
#    
#        plt.plot(data, label="function {}".format(funct))
#    plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
#    if norm:
#        plt.title("Normalized {}, detector={}, nod={}, order={}".format(star, detector, nod+1, order))
#    else:
#        plt.title("{}, detector={}, nod={}, order={}".format(star, detector, nod+1, order))
#    plt.show()
        norm_file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.norm.fits")   
        file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.fits")
    
        this_nod = file_names[nod]
        norm_nod = norm_file_names[nod]
        
        this_file = os.path.join(this_folder, this_nod)
        norm_file = os.path.join(this_folder, norm_nod)
       
        data = fits.getdata(this_file)
        norm_data = fits.getdata(norm_file)
        if len(data) == 3:
            data = data[0][0]
        if len(norm_data) == 3:
            norm_data = norm_data[0][0]
            
        plt.subplot(211)
        plt.plot(data, label="function {}".format(funct))
        plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
        plt.title("{}, chip={}, nod={}, order={}".format(star, detector, nod+1, order))
        plt.xlim([0,1024])
        
        plt.subplot(212)
        plt.plot(data, label="function {}".format(funct))
        plt.title("Normalized {}, chip={}, nod={}, order={}".format(star, detector, nod+1, order))
        plt.xlim([0,1024])

    if save:
        save_name = "poly_cycle-b_{0}_chip_{1}_nod_{2}_o{3}.".format(star, detector, nod+1, order)
        plt.savefig(os.path.join(Save_dir, save_name + '.png'), bbox_inches='tight')
        plt.savefig(os.path.join(Save_dir, save_name + '.pdf'), bbox_inches='tight')
    
    if show:
        plt.show()
    else:
        plt.close()
def plot_same_function_different_orders(star="HD167665-1a", detector=1, funct="legendre", order=1, norm=False, nod=0, show=True, save=False):
    """ For each object, nod, and function plot the 4 different order of extraction"""
    
    orders = [1,2,3,4]
    
    data_path = "/home/jneal/Phd/data/Crires/BDs-DRACS/{}-copy_to_find_artifacts/".format(star)
    
    fig = plt.figure()
    fig.subplots_adjust(hspace=0.5)
    for order in orders:
        folder = "{}_o{}".format(funct, order)
        this_folder = os.path.join(data_path, folder) + "/"
    
        norm_file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.norm.fits")   
        file_names = get_filenames(this_folder, "*_{}.nod.*".format(detector), "*.ms.fits")

        this_nod = file_names[nod]
        norm_nod = norm_file_names[nod]
        
        this_file = os.path.join(this_folder, this_nod)
        norm_file = os.path.join(this_folder, norm_nod)
       
        data = fits.getdata(this_file)
        norm_data = fits.getdata(norm_file)
        if len(data) == 3:
            data = data[0][0]
        if len(norm_data) == 3:
            norm_data = norm_data[0][0]
            
        plt.subplot(211)
        plt.plot(data, label="order {}".format(order))
        plt.legend(loc=0, bbox_to_anchor=(1.4, 0.9), ncol=1, fancybox=True, shadow=True)
        plt.title("{}, chip={}, nod={}, function={}".format(star, detector, nod+1, funct))
        plt.xlim([0,1024])
        
        plt.subplot(212)
        plt.plot(norm_data, label="order {}".format(order))
        plt.title("Normalized {}, chip={}, nod={}, function={}".format(star, detector, nod+1, funct))
        plt.xlim([0,1024])
        
    if save:
        save_name = "poly_cycle-c_{0}_chip_{1}_{3}_nod_{2}.".format(star, detector, nod+1, funct)
        plt.savefig(os.path.join(Save_dir, save_name + '.png'), bbox_inches='tight')
        plt.savefig(os.path.join(Save_dir, save_name + '.pdf'), bbox_inches='tight')
    
    if show:
        plt.show()
    else:
        plt.close()
```

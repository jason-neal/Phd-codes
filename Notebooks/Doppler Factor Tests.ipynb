# Doppler factor doppler shift

Plot difference between using the relativistic doppler factor and the doppler shift


```python
import numpy as np
from astropy.constants import c
import astropy.units as u
import matplotlib.pyplot as plt

wave = np.linspace(1000, 2000, 10000) * u.nanometer  # nm


RV = np.arange(-1000, 1000,20) * u.kilometer / u.second

doppler_factor = np.sqrt((1 + RV/c) / (1 - RV/c))

non_relativisitic_doppler = (1 + RV / c)
```

```python
plt.plot(RV, doppler_factor, ".", )
plt.plot(RV, non_relativisitic_doppler)
plt.xlabel("Velocity (km/s)")
plt.show()
```

```python
plt.plot(RV, doppler_factor - non_relativisitic_doppler)
plt.ylabel("Delta shift factor (relativistic-non_relativistic) ")
plt.xlabel("Velocity (km/s)")
plt.show()
```
